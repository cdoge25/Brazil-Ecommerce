```python
import pandas as pd
```

```python
df_payments = pd.read_csv('olist_order_payments_dataset.csv')
```

```python
df_payments['payment_value'] = df_payments['payment_value'].round(2)
```

```python
df_payments
```

```python
def get_pid(p):
    if p == "boleto":
        return "P1"
    elif p == "credit_card":
        return "P2"
    elif p == "debit_card":
        return "P3"
    elif p == "voucher":
        return "P4"
    else:
        return "P0"
```

```python
df_payments['payment_type_id'] = df_payments['payment_type'].apply(get_pid)
```

```python
df_payments.to_csv("olist_order_payments_dataset.csv")
```
